## 2. Statistical Overview

```python
def data_overview(df):
    """
    Function to provide an overview of the dataset.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.

    Returns:
    None: Prints the overview of the DataFrame.
    """
    num_rows = df.shape[0]
    num_columns = df.shape[1]
    num_categorical = df.select_dtypes(include=['object']).shape[1]
    num_numerical = df.select_dtypes(exclude=['object']).shape[1]
    num_instances = df["id"].nunique()

    # Print the overview
    print(f"Number of rows: {num_rows}")
    print(f"Number of columns: {num_columns}")
    print(f"Number of categorical columns: {num_categorical}")
    print(f"Number of numerical columns: {num_numerical}")
    print(f"Number of unique instances: {num_instances}")
```

## 3. Missing Values

```python
# Function to calculate missing value percentages
def missing_percentage(df, dataset_name):
    total = df.shape[0]
    missing_count = df.isnull().sum()
    missing_percent = (missing_count / total) * 100
    return missing_percent.to_frame(dataset_name)

# Calculate missing percentages for each dataset
r_missing_pct = missing_percentage(df_real, 'Real')
h_missing_pct = missing_percentage(df_drawn, 'Hand-drawn')
s_missing_pct = missing_percentage(df_sim, 'Simulated')

# Combine the percentages
missing_percentages = r_missing_pct.join(h_missing_pct).join(s_missing_pct)
missing_percentages = missing_percentages.rename_axis('Feature').reset_index()

# Display the missing percentages
print(missing_percentages)
```

```python
# print columns that all three datasets are all null 
# (i.e. the columns that have 100% missing values in all datasets)
all_null_columns = missing_percentages[missing_percentages['Real'] == 100]
print(all_null_columns.Feature.unique())     
```

## 5. Analysis of Cateegorical Variables

#### Labels

```python
# function to plot the distribution of the numerical features in the dataset
import matplotlib.pyplot as plt
import seaborn as sns

def plot_distribution_all(df_real, df_sim, df_drawn):
    plt.figure(figsize=(25, 10))
    # Plot for Real dataset
    plt.subplot(1, 3, 1)
    sns.barplot(x=df_real.index, y=df_real.values)
    plt.xlabel('Average Label')
    plt.ylabel('Number of Instances')
    plt.title('Distribution of Average Label Faults - Real')
    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)
    
    # Plot for Simulated dataset
    plt.subplot(1, 3, 2)
    sns.barplot(x=df_sim.index, y=df_sim.values)
    plt.xlabel('Average Label')
    plt.ylabel('Number of Instances')
    plt.title('Distribution of Average Label Faults - Simulated')
    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)
    
    # Plot for Drawn dataset
    plt.subplot(1, 3, 3)
    sns.barplot(x=df_drawn.index, y=df_drawn.values)
    plt.xlabel('Average Label')
    plt.ylabel('Number of Instances')
    plt.title('Distribution of Average Label Faults - Drawn')
    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()

def plot_distribution(df):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=df.index, y=df.values)
    plt.xlabel('Average Label')
    plt.ylabel('Number of Instances')
    plt.title('Distribution of Average Label Faults')
    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)
    plt.show()

    plt.figure(figsize=(15, 5))
```

```python
# group by id and get the avg label
id_label = df_real.groupby('id')['label'].mean().reset_index()['label'].value_counts()
id_label_s = df_sim.groupby('id')['label'].mean().reset_index()['label'].value_counts()
id_label_h = df_drawn.groupby('id')['label'].mean().reset_index()['label'].value_counts()


plot_distribution_all(id_label, id_label_s, id_label_h)

```

```python
# group by id and get the avg label
id_label_r = df_real.groupby('id')['label'].mean().reset_index()['label'].value_counts()

id_label_s = df_sim.groupby(['id', 'label'])['class'].mean().reset_index()['label'].value_counts()

id_label_d = df_drawn.groupby(['id', 'label'])['class'].mean().reset_index()['label'].value_counts()

plot_distribution_all(id_label_r, id_label_s, id_label_d)
```

## 6. Numerical Analysis

```python
# List of columns for histograms (numerical)
columns_to_plot = ['ABER-CKGL', 'P-JUS-CKGL', 'P-MON-CKP', 'P-TPT', 'T-TPT', 'QGL']

# Loop through columns and plot histograms for each dataset
for column in columns_to_plot:
    plt.figure(figsize=(15, 5))
    
    # Plot for Real dataset
    plt.subplot(1, 3, 1)
    df_real[column].dropna().hist(bins=30, alpha=0.7)
    plt.title(f'Real Data: {column}')
    
    # Plot for Simulated dataset
    plt.subplot(1, 3, 2)
    if column in df_sim.columns:
        df_sim[column].dropna().hist(bins=30, alpha=0.7, color='orange')
        plt.title(f'Simulated Data: {column}')
    
    # Plot for Drawn dataset
    plt.subplot(1, 3, 3)
    if column in df_drawn.columns:
        df_drawn[column].dropna().hist(bins=30, alpha=0.7, color='green')
        plt.title(f'Drawn Data: {column}')
    
    plt.tight_layout()
    plt.show()
```

## 7. Date time analysis


```python
# group by id and get the min and max timestamp 
id_timestamp = df_real.groupby('id')['timestamp'].agg(['min', 'max']).reset_index()
id_timestamp
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


# Convert to datetime format
id_timestamp['min'] = pd.to_datetime(id_timestamp['min'])
id_timestamp['max'] = pd.to_datetime(id_timestamp['max'])

# Calculate the duration
id_timestamp['duration'] = id_timestamp['max'] - id_timestamp['min']

# Plotting the Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through each event and plot a horizontal bar
for i, row in id_timestamp.iterrows():
    ax.barh(row['id'], row['duration'].total_seconds() / 3600, left=row['min'], height=0.5, color='skyblue', edgecolor='black')

# Use AutoDateLocator to prevent too many ticks being generated
ax.xaxis_date()
locator = mdates.AutoDateLocator(minticks=3, maxticks=10)  # Adjust the min/max ticks
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels

# Labeling and layout adjustments
plt.xlabel('Time')
plt.ylabel('Event ID')
plt.title('Event Intervals Gantt Chart')

# Adding gridlines for better readability
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Assuming you have the dataframe ready
id_timestamp['min'] = pd.to_datetime(id_timestamp['min'])
id_timestamp['max'] = pd.to_datetime(id_timestamp['max'])
id_timestamp['duration'] = id_timestamp['max'] - id_timestamp['min']

# Sort the dataframe by start time to assign positions incrementally
id_timestamp = id_timestamp.sort_values('min').reset_index(drop=True)

# Assign an incremental position for each event based on its order
id_timestamp['position'] = range(len(id_timestamp))

# Extract years from start times
id_timestamp['year'] = id_timestamp['min'].dt.year

# Get a list of unique years for which to create subplots
years = id_timestamp['year'].unique()

def plot_gantt_for_year(df_year, ax, year):
    for i, row in df_year.iterrows():
        ax.barh(row['position'], row['duration'].total_seconds() / 3600,
                left=row['min'], height=0.5, color='skyblue', edgecolor='black')
    ax.xaxis_date()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.xticks(rotation=45, ha='right')
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.set_title(f'Event Intervals in {year}')


# Create subplots for each year
fig, axes = plt.subplots(nrows=len(years), figsize=(12, 5 * len(years)))  # Adjust the subplot size based on number of years

for ax, year in zip(axes, years):
    df_year = id_timestamp[id_timestamp['year'] == year]
    plot_gantt_for_year(df_year, ax, year)
    ax.set_xlabel('Time')
    ax.set_ylabel('Event Number')

plt.tight_layout()
plt.show()
```
