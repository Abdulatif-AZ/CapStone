## 3.2 Data Types


```python
# Get the data types of each DataFrame
r_dtypes = r_df.dtypes.to_frame('Real').reset_index()
h_dtypes = h_df.dtypes.to_frame('Hand-drawn').reset_index()
s_dtypes = s_df.dtypes.to_frame('Simulated').reset_index()

# Merge the data types DataFrames on column names
dtype_df = r_dtypes.merge(h_dtypes, on='index', how='outer').merge(s_dtypes, on='index', how='outer')

# Rename the 'index' column to 'Column'
dtype_df = dtype_df.rename(columns={'index': 'Column'})

# Reorder columns
dtype_df = dtype_df[['Column', 'Real', 'Hand-drawn', 'Simulated']]

# Display the data types comparison
print(dtype_df)
```

All columns across the 3 dataset have the same data type

###  Downcasting the data types to avoid out  of memory issues

```python
def optimize_dtypes(df):
    for col in df.columns:
        col_type = df[col].dtype

        if col_type == 'int64' or col_type == 'int32':
            c_min = df[col].min()
            c_max = df[col].max()
            if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                df[col] = df[col].astype(np.int8)
            elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                df[col] = df[col].astype(np.int16)
            else:
                df[col] = df[col].astype(np.int32)
        elif col_type == 'float64' or col_type == 'float32':
            c_min = df[col].min()
            c_max = df[col].max()
            if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                df[col] = df[col].astype(np.float16)
            else:
                df[col] = df[col].astype(np.float32)
    return df

# Apply to your DataFrames
r_df = optimize_dtypes(r_df)
h_df = optimize_dtypes(h_df)
s_df = optimize_dtypes(s_df)
```

## 3.3 Statistical Summary 

```python
# Generate the statistical summaries for each dataset
r_desc = r_df.describe().transpose()
h_desc = h_df.describe().transpose()
s_desc = s_df.describe().transpose()

# Add a column to identify the dataset
r_desc['Dataset'] = 'Real'
h_desc['Dataset'] = 'Hand-drawn'
s_desc['Dataset'] = 'Simulated'

# Add an index to identify the features
r_desc['Feature'] = r_desc.index
h_desc['Feature'] = h_desc.index
s_desc['Feature'] = s_desc.index

# Combine the summaries into one DataFrame
combined_desc = pd.concat([r_desc, h_desc, s_desc], axis=0)

# Rearrange columns for better readability
combined_desc = combined_desc[['Feature', 'Dataset', 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]

# Pivot the DataFrame to compare datasets side by side
pivot_desc = combined_desc.pivot(index='Feature', columns='Dataset')

# Flatten the MultiIndex columns
pivot_desc.columns = ['_'.join(col).strip() for col in pivot_desc.columns.values]

# Reset index to turn 'Feature' back into a column
pivot_desc = pivot_desc.reset_index()

# Display the combined summary
print(pivot_desc)

```

## 3.4 Missing Values


```python
# Calculate the number of missing values in each dataset
r_missing = r_df.isnull().sum().to_frame('Real')
h_missing = h_df.isnull().sum().to_frame('Hand-drawn')
s_missing = s_df.isnull().sum().to_frame('Simulated')

# Combine the missing values data
missing_values = r_missing.join(h_missing, how='outer').join(s_missing, how='outer')

# Reset index to have 'Feature' column
missing_values = missing_values.rename_axis('Feature').reset_index()

# Display the missing values comparison
print(missing_values)
```

### 3.4.1 Missing Values %%


```python
# Function to calculate missing value percentages
def missing_percentage(df, dataset_name):
    total = df.shape[0]
    missing_count = df.isnull().sum()
    missing_percent = (missing_count / total) * 100
    return missing_percent.to_frame(dataset_name)

# Calculate missing percentages for each dataset
r_missing_pct = missing_percentage(r_df, 'Real')
h_missing_pct = missing_percentage(h_df, 'Hand-drawn')
s_missing_pct = missing_percentage(s_df, 'Simulated')

# Combine the percentages
missing_percentages = r_missing_pct.join(h_missing_pct).join(s_missing_pct)
missing_percentages = missing_percentages.rename_axis('Feature').reset_index()

# Display the missing percentages
print(missing_percentages)
```
